```python
import numpy as np
```

```python
A = np.array([
    [56.0,0.0,4.4,68.0],
    [1.2,104.0,52.0,8.0],
    [1.8,135.0,99.0,0.9]
])
```

```python
print(A)
```

```python
cal = A.sum(axis=0)
print(cal)
```

```python
perc = 100*A/cal
print(perc)
```
